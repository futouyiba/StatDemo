```python
import pandas as pd

# 读取 Excel 文件，指定数据行从第二行开始（跳过标题行）
backlogXlsxPath = 'E:\\DocsHDD\\FGame\\中鱼\\中鱼backlog.xlsx'
backlogSheetName = '#0221鱼对应饵'

backlogDf = pd.read_excel(backlogXlsxPath, sheet_name=backlogSheetName, header=1)

# 重命名列名，便于后续处理
backlogDf.columns = ['id', '鱼种', '饵种1', '喜好度1', '饵种2', '喜好度2', '饵种3', '喜好度3', '饵种4', '喜好度4', '饵种5', '喜好度5']

# 构建鱼种到饵种喜好度的映射
fish_to_bait = {}

for index, row in backlogDf.iterrows():
    fish_species = row['鱼种']
    bait_prefs = {}
    for i in range(1, 6):
        bait_type = row[f'饵种{i}']
        preference = row[f'喜好度{i}']
        if pd.notna(bait_type) and pd.notna(preference):
            bait_prefs[bait_type] = preference
    fish_to_bait[fish_species] = bait_prefs

# 构建饵种到鱼种喜好度的映射
bait_to_fish = {}

for fish_species, bait_prefs in fish_to_bait.items():
    for bait_type, preference in bait_prefs.items():
        if bait_type not in bait_to_fish:
            bait_to_fish[bait_type] = {}
        bait_to_fish[bait_type][fish_species] = preference

# 功能1：给定一个饵种，查询出有哪些鱼种对其有喜好度
def get_fish_for_bait(bait_type):
    return bait_to_fish.get(bait_type, {})

# 功能2：给定鱼种和饵种，查询出对应的喜好度
def get_preference(fish_species, bait_type):
    bait_prefs = fish_to_bait.get(fish_species, {})
    return bait_prefs.get(bait_type, None)

# 示例用法：

# 1. 给定一个饵种，查询有哪些鱼种对其有喜好度
bait = 'Bass Jig'
fish_preferences = get_fish_for_bait(bait)
print(f"对饵种 {bait} 有喜好度的鱼种:")
for fish, preference in fish_preferences.items():
    print(f"{fish}: 喜好度 {preference}")

# 2. 给定鱼种和饵种，查询对应的喜好度
fish_species = '大口鲈-成年'
bait_type = 'Bass Jig'
preference = get_preference(fish_species, bait_type)
if preference is not None:
    print(f"{fish_species} 对饵种 {bait_type} 的喜好度为 {preference}")
else:
    print(f"{fish_species} 对饵种 {bait_type} 没有喜好度数据")

print(fish_to_bait)
```
