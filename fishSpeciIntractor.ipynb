```python
import pandas as pd

# Load the csv file
csvFile = "sourceTables\鱼种场景投放与资源盘点.csv"

df = pd.read_csv(csvFile, encoding='utf-8', dtype=str)
# Step 1: Drop rows with missing '美术编号'
filtered_df = df.dropna(subset=['美术编号'])

# Step 2: Drop duplicates based on '美术编号'
distinct_df = filtered_df.drop_duplicates(subset='美术编号', keep='first')

# Step 3: Drop specified columns
columns_to_drop = ['经验（XP）', '鱼类图片1', '鱼类图片2', '鱼类图片3', '3D资源名称', '策划负责人']
cleaned_df = distinct_df.drop(columns=columns_to_drop)

# 重新整理“编号”这一列，从1开始编号
cleaned_df['编号'] = range(1, len(cleaned_df) + 1)


```

```python
# 将“编号”这一列设置为索引
cleaned_df.set_index('编号', inplace=True)
```

```python
# save df to a new csv file
cleaned_csv_file = "sourceTables\cleaned_fish_species.csv"
cleaned_df.to_csv(cleaned_csv_file, index=False, encoding='utf-8-sig')
print(f"Cleaned data saved to {cleaned_csv_file}")
```
