```python
import matplotlib.pyplot as plt
import numpy as np

# 配置参数
max_depth = 10  # 最大深度
surface_min_thickness = 1  # 表层最小厚度
surface_depth_ratio = 0.2  # 表层深度比例
bottom_min_thickness = 1  # 底层最小厚度
bottom_depth_ratio = 0.2  # 底层深度比例

# 生成弯曲的河床
x = np.linspace(0, 100, 500)
riverbed = max_depth - 2 * np.sin(x / 10) - np.cos(x / 5)

# 计算表层和底层的深度
surface_depth = np.minimum(surface_min_thickness, surface_depth_ratio * max_depth)
bottom_depth = max_depth - np.minimum(bottom_min_thickness, bottom_depth_ratio * max_depth)

# 创建图形
fig, ax = plt.subplots(figsize=(12, 6))

# 绘制河床
ax.plot(x, riverbed, color='black', label='Riverbed')

# 绘制表层区域
ax.fill_between(x, riverbed, riverbed - surface_depth, where=(riverbed - surface_depth >= 0), color='red', alpha=0.3, label='Surface Layer')

# 绘制底层区域
ax.fill_between(x, riverbed, riverbed - bottom_depth, where=(riverbed - bottom_depth >= 0), color='blue', alpha=0.3, label='Bottom Layer')

# 绘制中层区域
ax.fill_between(x, riverbed - surface_depth, riverbed - bottom_depth, where=(riverbed - surface_depth >= 0) & (riverbed - bottom_depth >= 0), color='green', alpha=0.3, label='Middle Layer')

# 设置图形属性
ax.set_ylim(max_depth + 2, -2)
ax.set_xlim(0, 100)
ax.set_xlabel('Horizontal Position')
ax.set_ylabel('Depth')
ax.set_title('Water Layers in a Curved Riverbed')
ax.legend()

# 显示图形
plt.show()
```
