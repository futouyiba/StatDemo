```python
# 设置 scale 值为 1 / (1 - probability_no_fish)，即 1 / (1 - 1/15) = 1 / (14/15)
probability_no_fish = 14 / 15
scale = 1 / (1 - probability_no_fish)

# 生成指数分布数据
data = np.random.exponential(scale=scale, size=10000)

# 绘制新的指数分布的直方图
plt.hist(data, bins=50, density=True, alpha=0.6, color='b')

# 添加标题和标签
plt.title("Exponential Distribution (P(fish)=1/15)")
plt.xlabel("Time to next fish")
plt.ylabel("Density")

plt.show()
```
