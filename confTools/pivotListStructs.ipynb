# 2001渠道表格

```python
# const define
workSpacePath = r".."
affExcelPath = r"..\configs\xlsx_channel\2001\fish_env_affinity.xlsx"
waterStructSheetName = 'StructAffinity'
pivotCsvPath = r"..\intermediateTables\structPivot.csv"
pivotXlsxPath = r"..\intermediateTables\structPivot.xlsx"

import pandas as pd
import numpy as np
import os
import sys
import openpyxl

# 第一行是数据类型，第二行是列名，第三行是说明和索引，第四行是wolong工具标记，第五行开始是数据
# 读取数据
affinityData = pd.read_excel(affExcelPath, sheet_name=waterStructSheetName, header=0,skiprows=[0,2,3])
affinityData
```

## PIVOT

```python
# Extract the structure type names from the data
struct_types = [
    affinityData['struct_type'].iloc[0],
    affinityData['struct_type.1'].iloc[0],
    affinityData['struct_type.2'].iloc[0],
    affinityData['struct_type.3'].iloc[0],
    affinityData['struct_type.4'].iloc[0]
]
print("Structure types:", struct_types)

# Pivot the data: rows are fish species, columns are struct types
pivoted_data = affinityData.pivot_table(
    index='name',
    values=['coeff', 'coeff.1', 'coeff.2', 'coeff.3', 'coeff.4'],
    aggfunc='first'
)
# Rename columns to use the actual structure type names
pivoted_data.columns = struct_types

pivoted_data.to_excel(pivotXlsxPath, index=True, header=True)
pivoted_data
```

```python
# Save the pivoted data to a CSV file
pivoted_data.to_csv(pivotCsvPath, index=True, header=True)
```

### Save 2 pivot EXCEL

```python
pivoted_data.to_excel(pivotXlsxPath, index=True, header=True)
```

## 反向UNPIVOT

```python
# 反向转换：把透视表转回原始表格结构并更新数据

# 创建原始数据的副本以保留其结构
updated_affinity_data = affinityData.copy()

# 遍历原始数据的每一行
for idx, row in updated_affinity_data.iterrows():
    fish_name = row['name']
    
    # 检查这个鱼种是否存在于我们的透视表中
    if fish_name in pivoted_data.index:
        # 从透视表获取系数值
        coeffs = pivoted_data.loc[fish_name].values
        
        # 更新原始数据结构中的系数列
        updated_affinity_data.at[idx, 'coeff'] = coeffs[0]    # [水下结构体]开放水域
        updated_affinity_data.at[idx, 'coeff.1'] = coeffs[1]  # [水下结构体]水草
        updated_affinity_data.at[idx, 'coeff.2'] = coeffs[2]  # [水下结构体]石头
        updated_affinity_data.at[idx, 'coeff.3'] = coeffs[3]  # [水下结构体]沉木
        updated_affinity_data.at[idx, 'coeff.4'] = coeffs[4]  # [水下结构体]桥墩

# 显示更新后的表格
updated_affinity_data
```

```python
# 将更新后的数据写回Excel文件，保留原格式
from openpyxl import load_workbook

def update_excel_with_dataframe(excel_path, sheet_name, df, start_row=5, start_col=1):
    """
    使用DataFrame的值更新Excel文件，同时保持原始格式。
    
    参数:
    excel_path (str): Excel文件的路径
    sheet_name (str): 要更新的工作表名称
    df (DataFrame): 包含更新数据的DataFrame
    start_row (int): 数据开始的行索引（默认为5，因为前4行有特殊含义）
    start_col (int): 数据开始的列索引（默认为1，即B列）
    """
    # 加载现有Excel文件
    print(f"正在加载Excel文件：{excel_path}")
    workbook = load_workbook(excel_path)
    worksheet = workbook[sheet_name]
    
    # 定义需要更新的列（系数列）
    coeff_cols = ['coeff', 'coeff.1', 'coeff.2', 'coeff.3', 'coeff.4']
    # 查找这些列在DataFrame中的索引位置
    col_indices = [df.columns.get_loc(col) for col in coeff_cols]
    
    # 更新Excel单元格
    print("正在更新Excel单元格...")
    for df_row_idx in range(len(df)):
        excel_row_idx = start_row + df_row_idx  # 计算Excel中的行索引
        
        # 只更新系数列
        for idx, col_name in enumerate(coeff_cols):
            # 系数列在Excel中的实际列位置（假设它们是按顺序排列的）
            excel_col_idx = start_col + col_indices[idx]
            
            # 获取DataFrame中的值
            value = df.iloc[df_row_idx][col_name]
            
            # 更新Excel单元格
            cell = worksheet.cell(row=excel_row_idx, column=excel_col_idx)
            cell.value = value
    
    # 保存更新后的Excel文件
    print(f"正在保存更新后的Excel文件：{excel_path}")
    workbook.save(excel_path)
    print("Excel文件已成功更新！")

# 调用函数更新Excel文件
# 注释掉下面一行以避免误操作覆盖原始文件
update_excel_with_dataframe(affExcelPath, waterStructSheetName, updated_affinity_data)
```

## 特殊结构Excel文件的读写

```python
# 用于读取和更新具有特殊结构的Excel文件（带有类型、中文名和标记行）
import pandas as pd
import openpyxl

class StructuredExcel:
    """处理带有特殊行的Excel文件：
       第1行: 数据类型（INT64, STRING等）
       第2行: 英文列名（作为DataFrame的列名）
       第3行: 中文列名
       第4行: 数据结构标记（[List]#1等）
       第5行+: 实际数据行
    """
    
    def __init__(self, excel_path):
        self.excel_path = excel_path
        self.workbook = None
        self._load_workbook()
    
    def _load_workbook(self):
        self.workbook = openpyxl.load_workbook(self.excel_path)
    
    def get_sheet_names(self):
        """获取所有工作表名称"""
        return self.workbook.sheetnames
    
    def read_data(self, sheet_name):
        """读取数据部分，返回DataFrame，跳过类型、中文名和标记行"""
        return pd.read_excel(self.excel_path, sheet_name=sheet_name, header=0, skiprows=[0, 2, 3])
    
    def read_metadata(self, sheet_name):
        """读取元数据行"""
        # 读取类型行（第1行）
        types_row = pd.read_excel(self.excel_path, sheet_name=sheet_name, header=None, nrows=1)
        # 读取中文名行（第3行）
        chinese_row = pd.read_excel(self.excel_path, sheet_name=sheet_name, header=None, skiprows=2, nrows=1)
        # 读取标记行（第4行）
        marker_row = pd.read_excel(self.excel_path, sheet_name=sheet_name, header=None, skiprows=3, nrows=1)
        
        return {
            'types': types_row.iloc[0].to_dict(),
            'chinese_names': chinese_row.iloc[0].to_dict(),
            'markers': marker_row.iloc[0].to_dict()
        }
    
    def update_cells(self, sheet_name, df, start_row=5, columns_to_update=None):
        """更新工作表中的特定单元格，保留原结构"""
        worksheet = self.workbook[sheet_name]
        
        # 如果没有指定要更新的列，则更新所有列
        if columns_to_update is None:
            columns_to_update = df.columns.tolist()
        
        # 查找列在DataFrame中的索引位置
        col_indices = [df.columns.get_loc(col) for col in columns_to_update if col in df.columns]
        
        print(f"正在更新工作表 '{sheet_name}' 中的数据...")
        for df_row_idx in range(len(df)):
            excel_row_idx = start_row + df_row_idx  # 计算Excel中的行索引
            
            # 更新指定列
            for idx, col_name in enumerate([columns_to_update[i] for i in range(len(col_indices))]):
                # 列在Excel中的索引位置（基于1）
                excel_col_idx = col_indices[idx] + 1
                
                # 获取DataFrame中的值
                value = df.iloc[df_row_idx][col_name]
                
                # 更新Excel单元格
                cell = worksheet.cell(row=excel_row_idx, column=excel_col_idx)
                cell.value = value
        
        print(f"保存更新后的Excel文件: {self.excel_path}")
        self.workbook.save(self.excel_path)
        print("Excel文件已成功更新！")

    def add_new_rows(self, sheet_name, df, start_row=None):
        """在工作表末尾添加新行"""
        worksheet = self.workbook[sheet_name]
        
        # 如果没有指定开始行，则从最后一行之后开始
        if start_row is None:
            start_row = worksheet.max_row + 1
        
        print(f"正在向工作表 '{sheet_name}' 中添加新行...")
        for df_row_idx in range(len(df)):
            excel_row_idx = start_row + df_row_idx
            
            # 添加所有列
            for col_idx, col_name in enumerate(df.columns):
                excel_col_idx = col_idx + 1
                value = df.iloc[df_row_idx][col_name]
                worksheet.cell(row=excel_row_idx, column=excel_col_idx).value = value
        
        print(f"保存更新后的Excel文件: {self.excel_path}")
        self.workbook.save(self.excel_path)
        print("Excel文件已成功更新！")

# 示例使用
# excel = StructuredExcel(affExcelPath)
# data_df = excel.read_data(waterStructSheetName)
# metadata = excel.read_metadata(waterStructSheetName)
# print("列的数据类型:", metadata['types'])
# print("列的中文名称:", metadata['chinese_names'])
# print("列的结构标记:", metadata['markers'])
# excel.update_cells(waterStructSheetName, updated_affinity_data, columns_to_update=['coeff', 'coeff.1', 'coeff.2', 'coeff.3', 'coeff.4'])
```
